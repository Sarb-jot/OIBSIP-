### Data Preprocessing

```python
df.columns= df.columns.str.strip()
```

```python
df1.columns= df1.columns.str.strip()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df1['Date'] = pd.to_datetime(df1['Date'])
```

### Data Visualizations

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x="Date",y="Estimated Unemployment Rate (%)",data=df1)
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.title("Unemployment Rate Over Time")
plt.grid(True)
plt.show()
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df1)
```

```python
fig = plt.figure(figsize = (10, 10))
sns.histplot(x='Estimated Unemployment Rate (%)', data=df, kde=True, hue='Area')
plt.title('Unemployment according to Area')
plt.xlabel('Unemployment Rate')
plt.show()
```

```python
fig = plt.figure(figsize = (5, 5))
sns.lineplot(y='Estimated Unemployment Rate (%)', x='Date', data=df)
plt.title('Unemployment according to Date')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
plt.figure(figsize = (30, 15))
plt.scatter(df1['Date'], df1['Estimated Employed'])

plt.title('Unemployment according to Region')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
fig = plt.figure(figsize = (30, 15))
sns.histplot(x='Estimated Labour Participation Rate (%)', data=df, kde=True, hue='Area')
plt.title('Labour Participation according to Area')
plt.xlabel('Labour Participation Rate')
plt.show()
```

```python
fig = plt.figure(figsize = (7, 7))
sns.lineplot(y='Estimated Labour Participation Rate (%)', x='Date', data=df)
plt.title('Labour Participation according to Date')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Labour Participation Rate')
plt.show()
```

```python
sns.histplot(x='Estimated Employed', data=df, kde=True, hue='Area')
plt.title('Employment according to Area')
plt.xlabel('Employment Rate')
plt.show()
```

```python
area_trend = df.groupby(['Date', 'Area'])['Estimated Unemployment Rate (%)'].mean().reset_index()

plt.figure(figsize=(14, 7))
sns.lineplot(data=area_trend, x='Date', y='Estimated Unemployment Rate (%)', hue='Area', marker='o')

plt.title('Unemployment Rate Over Time: Urban vs Rural')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend(title='Area')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Month_Year'] = df['Date'].dt.to_period('M').astype(str)

heatmap_data = df.pivot_table(index='Region', columns='Month_Year', values='Estimated Unemployment Rate (%)')

plt.figure(figsize=(16, 10))
sns.heatmap(heatmap_data, cmap='YlOrRd', annot=True, fmt=".1f", linewidths=0.5, linecolor='gray')

plt.title('Heatmap of Unemployment Rate by State and Month')
plt.xlabel('Month-Year')
plt.ylabel('State')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
